```python
import matplotlib.pyplot as plt
import numpy as np

# Definition the function to simulate die rolls and visualize the outcomes
def simulate_die_rolls(num_rolls=500):
    # Define the faces of the die
    faces = [1, 2, 3, 4, 5, 6]
    
    # Simulate the fair die (equal probabilities)
    fair_die_probs = [1/6] * 6
    fair_rolls = np.random.choice(faces, size=num_rolls, p=fair_die_probs)
    
    # Simulate the biased die (biased probabilities)
    biased_die_probs = [0.15, 0.15, 0.15, 0.15, 0.15, 0.25]
    biased_rolls = np.random.choice(faces, size=num_rolls, p=biased_die_probs)
    
    # Count the occurrences of each face
    fair_counts = np.bincount(fair_rolls, minlength=7)[1:]
    biased_counts = np.bincount(biased_rolls, minlength=7)[1:]
    
    # Plot the outcomes
    fig, ax = plt.subplots(1, 2, figsize=(12, 6), sharey=True)
    
    # Fair die plot
    ax[0].bar(faces, fair_counts, color='pink', alpha=0.7)
    ax[0].set_title(f'Fair Die (n={num_rolls})')
    ax[0].set_xlabel('Fair Die Face')
    ax[0].set_ylabel('Count')
    ax[0].set_xticks(faces)
    
    # Biased die plot
    ax[1].bar(faces, biased_counts, color='purple', alpha=0.7)
    ax[1].set_title(f'Biased Die (n={num_rolls})')
    ax[1].set_xlabel('Biased Die Face')
    ax[1].set_xticks(faces)
    
    plt.suptitle('Comparison of Fair and Biased Die Rolls')
    plt.tight_layout()
    plt.show()

# Run the function with the number of rolls
simulate_die_rolls()
```
